Buatlah visualisasi wordcloud dari kumpulan opini/komentar terkait topik spesifik tertentu berbahasa Indonesia/Inggris.

Note: File yang di-submit adalah file jupyter notebook yang digunakan untuk menghasilkan visualisasi wordcloud dan datanya.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

##Dataset

Dataset didapatkan di [kaggle](https://www.kaggle.com/shitalkat/amazonearphonesreviews?select=AllProductReviews.csv) tentang review produk earphone dengan bluetooth di situs belanja Amazon

```python
df = pd.read_csv('https://www.dropbox.com/s/1r1106j6dwovfv0/AllProductReviews.csv?dl=1')
df.head()
```

##Data exploration

```python
print('Ada {} baris data dengan {} atribut'.format(df.shape[0],df.shape[1]))
print('Ada {} produk unik yang memiliki review, yaitu antara lain {}'.format(len(df.Product.unique()),
                                                                 ", ".join(df.Product.unique()[:3]))
)
print('Rating terendah adalah {}, sedangkan rating tertinggi adalah {}'.format(min(df.ReviewStar),
                                                                        max(df.ReviewStar)))
```

```python
print('Deskripsi data yang bertipe data numeric:')
df.describe()
```

```python
text = " ".join(text for text in df.ReviewBody)
print ("Terdapat total {} kata dari semua review yang terdapat di reviewBody.".format(len(text)))
```

##Wordcloud tanpa stop words

```python
wordcloud = WordCloud(background_color="white").generate(text)

plt.figure(figsize=(16,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

##Wordcloud dengan stopwords

```python
stopwords = set(STOPWORDS)
stopwords.update(['now','got','still','much','sometime','look',
                  'day','come','really','high','using','brand',
                  'used','boat','need','while','year','month',
                  'day','keep','first','Amazon','design','want',
                  'hour','work'])
```

```python
wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)

plt.figure(figsize=(16,8))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
