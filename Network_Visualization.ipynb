##Import LIbrary

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

##Inisialisasi Node Kota dan Jalur Antar Kota

```python
node = [
        ['Bekasi',(2,4)],
        ['Karawang',(2.7,3.7)],
        ['Bogor',(1.25,3.5)],
        ['Sukabumi',(2,1.8)],
        ['Cianjur',(2.75,2.5)],
        ['Purwakarta',(3.5,3.25)],
        ['Bandung',(4.8,2.15)],
        ['Subang',(5,3.3)],
        ['Purwakarta',(3.5,3)],
        ['Cirebon',(10,3)],
        ['Sumedang',(6,2.7)],
        ['Tasikmalaya',(7.5,1.7)],
        ['Ciamis',(9,1.35)],
        ['Banjar',(10,1)],
        ['Garut',(6.3,1.5)],
        ['Majalengka',(8,2.4)],
        ['Indramayu',(9,3.75)]
]

path = [
        ('Bekasi','Bogor'),
        ('Bekasi','Karawang'),
        ('Bogor','Cianjur'),
        ('Bogor','Sukabumi'),
        ('Sukabumi','Cianjur'),
        ('Karawang','Purwakarta'),
        ('Karawang','Subang'),
        ('Karawang','Indramayu'),
        ('Purwakarta','Subang'),
        ('Purwakarta','Bandung'),
        ('Bandung','Sumedang'),
        ('Bandung','Cianjur'),
        ('Sumedang','Majalengka'),
        ('Sumedang','Cirebon'),
        ('Sumedang','Subang'),
        ('Cirebon','Indramayu'),
        ('Bandung','Garut'),
        ('Bandung','Tasikmalaya'),
        ('Tasikmalaya','Garut'),
        ('Tasikmalaya','Ciamis'),
        ('Ciamis','Banjar'),
        ('Indramayu','Cirebon')
]
```

##Graph Network

```python
G = nx.Graph()

for i in range(len(node)):
  G.add_node(node[i][0], pos=node[i][1])
  G.add_edges_from(path)

pos = nx.get_node_attributes(G,'pos')
plt.figure(figsize=(12,10))
nx.draw(G,pos,
        with_labels=True,
        node_color='royalblue',
        edge_color='cornflowerblue',
        node_size=600,
        width=2
)
```

##Degree Centrality

```python
nx.degree_centrality(G)
```

```python
degree_central_max = pd.DataFrame.from_dict({
    'node': list(nx.degree_centrality(G).keys()),
    'centrality':list(nx.degree_centrality(G).values())
})
sorted = degree_central_max.sort_values('centrality',ascending=False)

print('Kota terpenting berdasarkan degree centrality:',sorted['node'][:1].to_string(index=False))
```

##Betweenness Centrality

```python
nx.betweenness_centrality(G)
```

```python
betweenness_central_max = pd.DataFrame.from_dict({
    'node': list(nx.betweenness_centrality(G).keys()),
    'centrality':list(nx.betweenness_centrality(G).values())
})
sorted = betweenness_central_max.sort_values('centrality',ascending=False)

print('Kota terpenting berdasarkan betweenness centrality:',sorted['node'][:1].to_string(index=False))
```
